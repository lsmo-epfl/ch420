```python
# import packages

import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
data_path = '/Users/xiaoqi/Downloads/block0/exercise_2.dat'    # chagne it to yours
data = np.loadtxt(data_path)
```

```python
# make plots
plt.figure(figsize=(8, 6))

plt.plot(data[:, 0], data[:, 1], c='blue', linestyle='--' , label="sin(x)")
plt.plot(data[:, 0], data[:, 2], c='orange', label='cos(x)')

plt.title("Trigonometric functions", fontdict={'fontsize': 15})
plt.xlabel(f"x(\N{DEGREE SIGN})", fontdict={'fontsize': 12})
plt.ylabel("y", fontdict={'fontsize': 12})

plt.legend(loc='best', fontsize=13)

plt.savefig("trigonometric.jpg")
```
