```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("data.dat")
x = data[:,0]
y = data[:,1]

plt.plot(x, y, label="V(r)", color="red")
plt.title("Lennard-Jones Potential")
plt.xlabel("r/sigma")
plt.ylabel("V/epsilon")
plt.grid()
plt.legend()
plt.tight_layout()
plt.savefig("plot.png")
plt.show()
plt.close()
```
