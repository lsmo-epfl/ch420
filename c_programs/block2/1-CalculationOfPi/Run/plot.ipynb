```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data (number of cycles vs relative error)
# hint: you can collect this in a .dat file by copying the output of the program and saving it
# you can try many times with different number of cycles to get the average
# just keep the l/d ratio the same throughout to have a fair comparison
data_1 = np.loadtxt('results.dat')
x_1 = data_1[:, 0]
y_1 = data_1[:, 1]
```

```python
# Plot the data
plt.scatter(x_1, y_1, label='error', linewidth=3)
# use log axis
plt.yscale('log')
plt.xscale('log')

plt.xlabel('Number of cycles')
plt.ylabel('Relative error')
plt.title('Relative error as a function of the number of cycles')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig('error.png')
plt.show()
plt.close()
```
