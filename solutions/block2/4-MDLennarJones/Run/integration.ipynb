```python
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
drift_files = sorted(glob("drift-0*.dat"))
temperatures = []
drifts = []
for file in drift_files:
    drift = np.loadtxt(file)
    drifts.append(drift)
    
    temp = float(file.split("drift-")[1].split(".dat")[0])
    temperatures.append(temp)
```

```python
velverlet_files = sorted(glob("drift-velocity*.dat"))
velverlet_drifts = []
for file in velverlet_files:
    drift = np.loadtxt(file)
    velverlet_drifts.append(drift)
```

```python
euler_files = sorted(glob("drift-euler*.dat"))
euler_drifts = []
for file in euler_files:
    drift = np.loadtxt(file)
    euler_drifts.append(drift)
```

```python
# plot for temp=0.7

plt.plot(drifts[-1][:, 0], np.log(drifts[-1][:, 1]), label="Verlet", marker="x")
plt.plot(velverlet_drifts[-1][:, 0], np.log(velverlet_drifts[-1][:, 1]), label="Velocity verlet", marker="x")
plt.plot(euler_drifts[-1][:, 0], np.log(euler_drifts[-1][:, 1]), label="Euler", marker="x")

plt.xlabel("Time step")
plt.ylabel("log(Delta U)")
plt.legend()
```
