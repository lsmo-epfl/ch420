```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data (5000 energy levels, varying temperature)
folder = 'data/3/'

data_1 = np.loadtxt(folder+'results_5000_1.dat')
x_1 = data_1[:, 0]
y_1 = data_1[:, 1]

data_2 = np.loadtxt(folder+'results_5000_2.dat')
x_2 = data_2[:, 0]
y_2 = data_2[:, 1]

data_3 = np.loadtxt(folder+'results_5000_3.dat')
x_3 = data_3[:, 0]
y_3 = data_3[:, 1]

data_10 = np.loadtxt(folder+'results_5000_10.dat')
x_10 = data_10[:, 0]
y_10 = data_10[:, 1]

data_100 = np.loadtxt(folder+'results_5000_100.dat')
x_100 = data_100[:, 0]
y_100 = data_100[:, 1]
```

```python
plt.figure()
plt.plot(x_1,y_1, lw=3, label='T=1', color='r')
plt.plot(x_2,y_2, lw=3, label='T=2', color='g')
plt.plot(x_3,y_3, lw=3, label='T=3', color='b')
plt.plot(x_10,y_10, lw=3, label='T=10', color='y')
plt.plot(x_100,y_100, lw=3, label='T=100', color='m')
plt.xlim(0, 20)
plt.xlabel('energy level')
plt.ylabel('distribution')
plt.title('Boltzmann distribution')
plt.legend()
plt.tight_layout()
plt.savefig('plots/boltzmann_distribution_3.png')
plt.show()
plt.close()
```
