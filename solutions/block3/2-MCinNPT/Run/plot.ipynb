```python
import numpy as np
def extract_pressure(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    for line in lines:
        if "Fraction Succes (Volume)" in line:
            avg_pressure = float(line.split(":")[1].strip())

    return avg_pressure

# List of values
values_str = ['0.0001', '0.0005', '0.001', '0.005', '0.01', '0.05', '0.1', '0.5', '1', '5', '10', '50']
values = [float(value) for value in values_str]
values_log = np.log(np.array(values))
pressure_ls = []
# Generate file names and use the function for each file
for value in values_str:
    file = f'output_{value}.txt'
    accept_p = extract_pressure(file)
    pressure_ls.append(accept_p)

import matplotlib.pyplot as plt
plt.plot(values_log, pressure_ls, 'o')
plt.xlabel('Log10(delta_Vmax)')
plt.ylabel('aceept ratio')
```
